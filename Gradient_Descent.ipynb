```python
import numpy as np
from sklearn.datasets import make_regression
```

```python
X,y = make_regression(n_samples = 4, n_features=1, n_informative=1, n_targets=1, noise = 80, random_state=13)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(X,y)
```

```python
#Applying OLS
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color = 'red')
```

```python
#Applying Gradient Descent assuming slope is constant i.e. m = 78.35
#assuming the starting value for intercept b = 0

y_pred = ((78.35 * X) + 0).reshape(4)
print(y_pred)
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color = 'red', label = 'OLS')
plt.plot(X, y_pred, color= '#00a65a', label = 'b = 0')
plt.legend()
```

```python
m = 78.35
b = 0

loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
#let's assume learning rate as 0.1
lr = 0.1

step_size = lr*loss_slope
step_size
```

```python
b = b - step_size
b
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)
print(y_pred1)
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color = 'red', label = 'OLS')
plt.plot(X, y_pred1, color= '#00a65a', label = 'b = {}'.format(b))
plt.plot(X, y_pred, color= '#A3E4D7', label = 'b = 0')
plt.legend()
```

```python
# Iteration2
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35 * X) + b).reshape(4)
y_pred2
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color = 'red', label = 'OSL')
plt.plot(X, y_pred, color = '#00a65a', label = 'b = 0')
plt.plot(X, y_pred1, color = '#a3e4d7', label = 'b = {}'.format(b))
plt.plot(X, y_pred2, color = 'blue', label = 'b = {}'.format(b))
plt.legend()
```

```python
#Iteration
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = lr*loss_slope
```

```python
print(step_size)
```

```python
b = b - step_size
b
```

```python
y_pred3 = ((78.35 * X) + b).reshape(4)

plt.figure(figsize=(12,12))
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color = 'red', label = 'OSL')
plt.plot(X, y_pred, color = '#00a65a', label = 'b = 0')
plt.plot(X, y_pred1, color = '#00a65a', label = 'b = 20.927763408209216')
plt.plot(X, y_pred2, color = 'pink', label = 'b = 25.11331608985106')
plt.plot(X, y_pred3, color = 'blue', label = 'b = {}'.format(b))
plt.legend()
```

```python
b = -100
n = 78.35
lr = 0.1

epochs = 10

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*X.ravel() - b)
  b = b - (lr * loss_slope)

  y_pred = n * X + b

  plt.plot(X, y_pred)

plt.scatter(X,y)
```
